```python
df['Growth Rate'] = df['Growth Rate'].astype(str).str.rstrip('%').astype(float)

# Drop NaNs
df = df.dropna(subset=['Growth Rate'])

# Calculate IQR
Q1 = df['Growth Rate'].quantile(0.25)
Q3 = df['Growth Rate'].quantile(0.75)
IQR = Q3 - Q1

# Determine outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Growth Rate'] < lower_bound) | (df['Growth Rate'] > upper_bound)]

print(f"Number of Outliers: {len(outliers)}")
print("Outlier Countries:\n", outliers[['Country/Territory', 'Growth Rate']])

# ✅ Boxplot for Visualization
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Growth Rate'], color='skyblue')
plt.title('Outliers in Population Growth Rate')
plt.xlabel('Growth Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
from scipy.stats import ttest_ind, chi2_contingency
import pandas as pd

# T-test: Compare mean 2022 Population between Asia and Europe
group1 = df[df['Continent'] == 'Asia']['2022 Population']
group2 = df[df['Continent'] == 'Europe']['2022 Population']

t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
print("T-test: 2022 Population between Asia and Europe")
print(f"t-statistic: {t_stat:.3f}, p-value: {p_value:.3f}\n")

# Chi-Square Test: Association between Continent and Capital
contingency_table = pd.crosstab(df['Continent'], df['Capital'])

chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print("Chi-Square Test: Continent vs Capital")
print(f"chi2-statistic: {chi2_stat:.3f}, p-value: {p_val:.3f}, degrees of freedom: {dof}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Year-wise population columns
year_columns = [
    '1970 Population', '1980 Population', '1990 Population',
    '2000 Population', '2010 Population', '2015 Population',
    '2020 Population', '2022 Population'
]

# Empty lists to store results
years = []
means = []
medians = []
modes = []

# Calculate stats for each year
for col in year_columns:
    years.append(col.split()[0])  # just the year part
    means.append(df[col].mean())
    medians.append(df[col].median())
    modes.append(df[col].mode()[0])  # first mode value if multiple

# Create a DataFrame for plotting
stats_df = pd.DataFrame({
    'Year': years,
    'Mean': means,
    'Median': medians,
    'Mode': modes
})

# Plot the stats
plt.figure(figsize=(12, 6))
plt.plot(stats_df['Year'], stats_df['Mean'], label='Mean', marker='o')
plt.plot(stats_df['Year'], stats_df['Median'], label='Median', marker='s')
plt.plot(stats_df['Year'], stats_df['Mode'], label='Mode', marker='^')

plt.title('Mean, Median, and Mode of Population Over the Years')  # ✅ No emoji
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
# Prepare year and population lists
years = [int(col.split()[0]) for col in year_columns]
global_population = df[year_columns].sum().values  # Total population per year

# Convert to numpy arrays and reshape for model
X = np.array(years).reshape(-1, 1)
y = np.array(global_population).reshape(-1, 1)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict future years
future_years = np.array([2025, 2030, 2040]).reshape(-1, 1)
predictions = model.predict(future_years)

# Print predictions
for year, pop in zip(future_years.flatten(), predictions.flatten()):
    print(f"Predicted Global Population in {year}: {pop:,.0f}")
```

```python
# Combine actual and future for plotting
all_years = np.append(X.flatten(), future_years.flatten())
all_pops = np.append(y.flatten(), predictions.flatten())

plt.figure(figsize=(10, 6))
plt.plot(X, y, 'bo-', label='Actual Population')
plt.plot(future_years, predictions, 'ro--', label='Predicted Population')
plt.xlabel('Year')
plt.ylabel('Global Population')
plt.title('Global Population Prediction using Linear Regression')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
pop_columns = [
    '1970 Population', '1980 Population', '1990 Population',
    '2000 Population', '2010 Population', '2015 Population',
    '2020 Population', '2022 Population'
]
# Drop rows with missing population data
df_cluster = df.dropna(subset=pop_columns)

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cluster[pop_columns])
```

```python
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method to Determine Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
sns.scatterplot(
    x=df_cluster['2020 Population'],
    y=df_cluster['2022 Population'],
    hue=df_cluster['Cluster'],
    palette='Set1'
)

plt.title('K-Means Clustering of Countries by Population')
plt.xlabel('2020 Population')
plt.ylabel('2022 Population')
plt.grid(True)
plt.tight_layout()
plt.show()
```
